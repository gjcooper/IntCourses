```python
import requests
from scrapy.http import TextResponse

res = requests.get('https://www.aph.gov.au/Senators_and_Members/Parliamentarian_Search_Results')
response = TextResponse(res.url, body=res.text, encoding='utf-8')
```

```python
[l.splitlines() for l in response.xpath("//h4[@class='title']/..").extract()]
```

```python
for resp in response.xpath("//h4[@class='title']/.."):
    print(resp.xpath("h4/a/text()").extract())
    print(resp.xpath("dl/dd/text()").extract()[0:2])
```
